# Normalization:

```python
y = []
norm_fac_y = max(raw_y) - min(raw_y)
mean_y = sum(raw_y) / len(raw_y)
for i in raw_y:
    i = (i - mean_y) / norm_fac_y
    y.append(i)
y = np.array(y)

x = []
norm_fac_x = max(raw_x) - min(raw_x)
mean_x = sum(raw_x) / len(raw_x)
for i in raw_x:
    i = (i - mean_x) / norm_fac_x
    x.append(i)
x = np.array(x)
# '''
complete_y = []
for i in complete_raw_y:
    i = (i - mean_y) / norm_fac_y
    complete_y.append(i)
complete_y = np.array(y)

complete_x = []
for i in complete_raw_x:
    i = (i - mean_x) / norm_fac_x
    complete_x.append(i)
complete_x = np.array(complete_x)
# '''
```

# Gradient Descent:

```python
m = x.shape[0]
dj_da = 1
dj_db = 1
dj_ds = 1
dj_dt = 1
a = 1
b = 1
s = 1
t = 1
arrC = []
arrA = []
arrB = []
arrS = []
arrT = []
for j in range(100000):
    dj_da = 0
    dj_db = 0
    dj_ds = 0
    dj_dt = 0
    cost_sum = 0
    for i in range(m):
        temp = np.sqrt(abs(a* x[i] + b))
        f_abef = s * temp + t
        # f_wb = w * x[i] + b
        cost = (f_abef - y[i]) ** 2
        cost_sum += cost
        # dj_dw_i = (f_wb - y[i]) * x[i]
        dj_da_i = (f_abef - y[i]) * (s * x[i]) / (2 * temp)
        # dj_db_i = f_wb - y[i]
        dj_db_i = (f_abef - y[i]) * s / (2 * temp)
        dj_ds_i = (f_abef - y[i]) * temp
        dj_dt_i = f_abef - y[i]
        dj_da += dj_da_i
        dj_db += dj_db_i
        dj_ds += dj_ds_i
        dj_dt += dj_dt_i
    total_cost = (1 / (2 * m)) * cost_sum
    arrC.append(total_cost)
    # print(total_cost)
    # learning rate alpha is handled manually
    alpha = 0.1
    # w = w - (a * (dj_dw / m))
    # arrW.append(w)
    # b = b - (a * (dj_db / m))
    # arrB.append(b) m
    a = a - (alpha * (dj_da / m))
    arrA.append(a)
    b = b - (alpha * (dj_db / m))
    arrB.append(b)
    s = s - (alpha * (dj_ds / m))
    arrS.append(s)
    t = t - (alpha * (dj_dt / m))
    arrT.append(t)
    if (-0.3 < dj_da < 0.3) and (-0.3 < dj_db < 0.3) and (-0.3 < dj_ds < 0.3) and (-0.3 < dj_dt < 0.3):
        break

a_final = a
b_final = b
s_final = s
t_final = t
print("a:", a_final)
print("b:", b_final)
print("s:", s_final)
print("t:", t_final)
```

#  Cost Function and Learning Rate

```python
p = list(range(len(arrC)))
print("number of iterations:", len(p))
plt.title('cost function')
plt.xticks(np.arange(0, len(p), 20))
plt.xlabel('iteration no ---->')
plt.ylabel('function cost ---->')
plt.plot(p, arrC, c= 'r')
plt.show()

plt.title('learning rate of a')
plt.xlabel('value of a ---->')
plt.ylabel('function cost ---->')
plt.plot(arrA, arrC, c = 'y')
plt.show()

plt.title('learning rate of b')
plt.xlabel('value of b ---->')
plt.ylabel('function cost ---->')
plt.plot(arrB, arrC, c = 'b')
plt.show()

plt.title('learning rate of s')
plt.xlabel('value of s ---->')
plt.ylabel('function cost ---->')
plt.plot(arrS, arrC, c = 'b')
plt.show()

plt.title('learning rate of t')
plt.xlabel('value of t ---->')
plt.ylabel('function cost ---->')
plt.plot(arrT, arrC, c = 'b')
plt.show()
```

# Predictions:

```python
# y_predict = np.dot(w_final, complete_x) + b_final
# y_predict = np.dot(w_final, x) + b_final
temp = np.sqrt(abs(np.dot(a, complete_x) + b))
# temp = np.sqrt(abs(np.dot(a, x) + b))
y_predict = np.dot(s, temp) + t 
raw_y_predict = (y_predict * norm_fac_y) + mean_y
```
