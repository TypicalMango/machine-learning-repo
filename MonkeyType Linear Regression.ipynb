```python
import numpy as np
import matplotlib.pyplot as plt
```

# Raw Inputs:

```python
raw_y = [61.56, 58.78, 70.38, 79.20, 74.38, 71.55, 66.78, 79.60, 64.40,
         74.35, 71.98, 69.58, 77.97, 67.98, 72.80, 69.58, 77.95, 65.58,
         76.40, 69.98, 63.96, 85.57, 80.35, 85.60, 79.57, 81.57, 81.17,
         73.58, 82.40, 79.57, 81.97, 71.98, 81.97, 81.17, 80.75, 79.55,
         91.54, 78.77, 85.94, 73.18, 64.00, 84.00, 85.54, 77.57, 73.58,
         79.97, 78.77, 79.95, 82.40, 77.17, 83.20, 73.98, 76.75, 77.57,
         94.00, 81.17, 73.98, 81.97, 89.17, 76.80, 75.20, 74.35, 83.14,
         85.14, 88.00, 73.60, 84.77, 78.77, 71.98, 77.17, 83.97, 85.57,
         89.17]

raw_x = list(range(len(raw_y)))
```

# Normalization:

```python
y = []
norm_fac_y = max(raw_y) - min(raw_y)
mean_y = sum(raw_y) / len(raw_y)
for i in raw_y:
    i = (i - mean_y) / norm_fac_y
    y.append(i)
y = np.array(y)

x = []
norm_fac_x = max(raw_x) - min(raw_x)
mean_x = sum(raw_x) / len(raw_x)
for i in raw_x:
    i = (i - mean_x) / norm_fac_x
    x.append(i)
x = np.array(x)
```

# Gradient Descent:

```python
m = x.shape[0]
dj_dw = 1
dj_db = 1
w = 0
b = 0
arrC = []
arrW = [0]
arrB = [0]
for j in range(10000):
    dj_dw = 0
    dj_db = 0
    cost_sum = 0
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i]) ** 2
        cost_sum += cost
        dj_dw_i = (f_wb - y[i]) * x[i]
        dj_db_i = f_wb - y[i]
        dj_db += dj_db_i
        dj_dw += dj_dw_i
    total_cost = (1 / (2 * m)) * cost_sum
    arrC.append(total_cost)
    # print(total_cost)
    # learning rate alpha is handled manually
    a = 0.1
    if -0.3 < dj_dw < 0.3 and -0.3 < dj_db < 0.3:
        break
    w = w - (a * (dj_dw / m))
    arrW.append(w)
    b = b - (a * (dj_db / m))
    arrB.append(b)

w_final = w
b_final = b
print("weight:", w_final)
print("bias:", b_final)
```

#  Cost Function and Learning Rate

```python
p = list(range(len(arrC)))

plt.title('cost function')
plt.xticks(np.arange(0, len(p), 20))
plt.xlabel('iteration no ---->')
plt.ylabel('function cost ---->')
plt.plot(p, arrC, c= 'r')
plt.show()

plt.title('learning rate of weight (w)')
plt.xlabel('value of w ---->')
plt.ylabel('function cost ---->')
plt.plot(arrW, arrC, c = 'y')
plt.show()

plt.title('learning rate of bias (b)')
plt.xlabel('value of b ---->')
plt.ylabel('function cost ---->')
plt.plot(arrB, arrC, c = 'b')
plt.show()
```

# Predictions:

```python
y_predict = np.dot(w_final, x) + b_final
raw_y_predict = (y_predict * norm_fac_y) + mean_y
y_predict = np.array(y_predict)
```

```python
plt.figure(figsize=(20, 10))
plt.xlabel('attempt no ---->')
plt.ylabel('typing speed (wpm) ---->')
plt.xticks(np.arange(0, len(raw_y), 5))
plt.yticks(np.arange(0, 100, 10))
plt.xlim(0, len(raw_y))
plt.ylim(0, 100)
plt.scatter(raw_x, raw_y, c='black', s=5)
plt.plot(raw_x, raw_y, c='black')
plt.plot(raw_x, raw_y_predict, linestyle='dashed', c='b')
plt.show()
```
